```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.violinplot(data=tips,y="tip")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(data=tips, x='sex', y='total_bill')
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.violinplot(data=tips,x="day",y="tip",hue="sex")
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.violinplot(data=tips,x="day",y="tip",hue="sex", split=True)
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.violinplot(data=tips,x="day",y="tip",hue="sex", inner="quartile",cut=0)
plt.show()
```
