```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
data_set =pd.read_csv("road_traffic_accidents_with_gender.csv")
```

```python
data_set.head()
```

# added a new column named total victim

```python
data_set["Total_victim"]=data_set["Number_Of_Killed"]+data_set["Number_Of_Injured"]
```

```python
data_set.head()
```

# finding total victim by location

```python
total_victim_by_location = data_set.groupby(["Location", "gender"])["Total_victim"].sum().reset_index()
total_victim_by_location=total_victim_by_location.sort_values(by="Total_victim")
print(total_victim_by_location)
plt.figure(figsize=(10,8))
plot=sns.barplot(data=total_victim_by_location, x="Location",y="Total_victim",hue="gender")

for container in plot.containers:
    plot.bar_label(container, padding=3)  

plt.title("Number of accident by division")
plt.xticks(rotation=30)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
plot = sns.barplot(data=data_set, x="Location", y="Total_victim", hue="gender",palette="Set3")

#for container in plot.containers:
    #plot.bar_label(container, padding=3)

plt.title("Accident Victims by Division and Gender (with Error Bars)")
plt.xticks(rotation=30)
plt.ylabel("Average Total Victims")
plt.tight_layout()
plt.show()
```
