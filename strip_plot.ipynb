```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
sns.stripplot(data=data,y="total_bill", jitter=0.3)
plt.show()
```

```python
sns.stripplot(data=data,y="total_bill",x="day", jitter=True,)
plt.show()
```

```python
sns.stripplot(data=data,y="total_bill", x= "day", jitter=True, hue="sex")
plt.show()
```

# strip plot whit violin plot

```python
sns.stripplot(data=data,x="day",y="total_bill",jitter=True)
sns.violinplot(data=data,x="day",y="total_bill")
plt.show()
```
