```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df= sns.load_dataset("iris")
```

```python
df.head()
```

```python
sns.set_style("darkgrid")
sns.pairplot(data=df)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.pairplot(data=df, hue="species")
plt.show()
```

```python
sns.set_style("darkgrid")
sns.pairplot(data=df,hue="species", palette="cool",markers="s")
plt.show()
```

```python

sns.pairplot(data=df,hue="species", palette="cool",markers="s", kind="reg")
plt.show()
```

# plot with specific variables

```python
sns.pairplot(data=df,hue="species", palette="cool",markers="s", kind="reg",x_vars=["sepal_length","sepal_width","petal_width"],y_vars=["petal_length"])
plt.show()
```

```python
sns.pairplot(data=df, corner=True)
plt.show()
```
