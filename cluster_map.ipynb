```python
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
glue = sns.load_dataset("glue")
pivoted = glue.pivot(index="Model", columns="Task", values="Score")

sns.clustermap(pivoted, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("GLUE Model Performance Clustermap")
plt.show()
```
