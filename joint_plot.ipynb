```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names())
```

```python
df = sns.load_dataset("penguins")
df=df.dropna()
df.head()
```

```python
sns.jointplot(data=df,x="bill_length_mm",y="flipper_length_mm")
```

```python

sns.jointplot(data=df,x="bill_length_mm",y="flipper_length_mm",hue="sex")
plt.show()
```

```python
sns.jointplot(data=df,x="bill_length_mm",y="flipper_length_mm", joint_kws={"s": 80, "marker": "D", "alpha": 0.5, "color":"c"}, marginal_kws={"color":"skyblue"})
plt.show()
```

# plot KDE plot on top of the joint plot

```python
joint_plot=sns.jointplot(data=df,x="bill_length_mm",y="flipper_length_mm", joint_kws={"s": 80, "marker": "D", "alpha": 0.5, "color":"c"}, marginal_kws={"color":"skyblue"})

joint_plot.plot_joint(sns.kdeplot)
plt.show()
```

```python

# Create a jointplot object
joint_plot = sns.jointplot(
    data=df,
    x="bill_length_mm",
    y="flipper_length_mm",
    kind="scatter",  # Keep the base plot
    joint_kws={
        "s": 80,              # Dot size
        "marker": "D",        # Diamond marker
        "alpha": 0.5,         # Transparency
        "color": "c"          # Cyan color
    },
    marginal_kws={
        "fill": True,         # Fill KDE area
        "color": "skyblue",   # Marginal KDE color
        "alpha": 0.3          # Marginal transparency
    },
    height=8                 # Bigger plot size
)

# Overlay a smooth KDE on the scatter area
joint_plot.plot_joint(sns.kdeplot, color="gray", linewidth=1, alpha=0.7)

plt.show()
```
