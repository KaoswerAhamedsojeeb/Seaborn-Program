```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
```

```python
df.head()
```

# simple displot

```python
sns.displot(data=df,x="total_bill")
plt.show()
```

# displot whit kernel density estimate or kdeplot and rugplot

```python
sns.displot(data=df,x="total_bill",kde=True, rug=True)
plt.show()
```

```python
sns.displot(data=df,x="total_bill",kde=True, rug=True, hue="time")
plt.show()
```

# whit column

```python
sns.displot(data=df,x="total_bill",kde=True, rug=True, hue="sex", col="time")
plt.show()
```

# with column and row

```python
sns.displot(data=df,x="total_bill",kde=True, rug=True, hue="sex", col="day", multiple="stack")
plt.show()
```

```python
sns.displot(data=df,x="tip",kde=True, rug=True, col="sex",row="day")
plt.show()
```

# displot whit ecdf plot

```python
sns.displot(data=df,x="tip",kind="ecdf", col="sex",row="day")
plt.show()
```
