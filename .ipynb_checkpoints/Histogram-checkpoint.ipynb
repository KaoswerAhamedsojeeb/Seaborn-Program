```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips= sns.load_dataset("tips")
tips.head(10)
```

```python
sns.histplot(data=tips,x="total_bill")
plt.title("Total bill distribution")
plt.ylabel("count")
plt.xlabel("total bill")
plt.show()
```

```python
sns.histplot(data=tips,x="total_bill",bins=10,color="skyblue", kde= "true")
#kde refers to the curve for a smooth distribution
plt.title("Total bill distribution")
plt.ylabel("count")
plt.xlabel("total bill")
plt.show()
```

```python


sns.histplot(data=tips,x="total_bill",hue="sex",element="step",stat="density",kde="true")
plt.xticks(np.arange(0,55,5))
plt.title("Total bill distribution")
plt.xlabel("total bill with sex")
plt.ylabel("count")
plt.show()
```

```python
bin_edges = np.arange(0,55,5)
bin_center = (bin_edges[:-1] + bin_edges[1:])
#print(bin_center)

sns.histplot(data=tips,x="total_bill",hue="sex",element="step",stat="density",kde="true",bins=bin_edges)
plt.xticks(bin_center)
plt.xlim(0,60)
plt.title("Total bill distribution")
plt.xlabel("total bill with sex")
plt.ylabel("count")
plt.show()
```

# finding probability

```python
bin_edges = np.arange(0,55,5)

sns.histplot(data=tips,x="total_bill",stat="probability",bins=bin_edges)
plt.xticks(bin_edges)
plt.xlim(0,60)
plt.title("Total bill distribution")
plt.xlabel("total bill with sex")
plt.ylabel("count")
plt.show()
```
