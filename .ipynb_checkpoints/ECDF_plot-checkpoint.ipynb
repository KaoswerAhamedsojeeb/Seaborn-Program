```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = sns.load_dataset("tips")
```

```python
dataset.head()
```

```python
sns.ecdfplot(data=dataset)
plt.show()
```

```python
sns.ecdfplot(data=dataset,x="total_bill")
plt.show()
```

```python
sns.ecdfplot(data=dataset,x="total_bill",hue="day")
plt.show()
```
