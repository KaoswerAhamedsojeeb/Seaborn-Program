```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

# all builtin data set of seaborn

```python
print(sns.get_dataset_names())
```

```python
tip=sns.load_dataset("tips")
tip.head()
```

```python
plt.figure(figsize=(8,6))
sns.boxenplot(data=tip,y="total_bill", x="day")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxenplot(data=tip,y="total_bill", x="day", palette='dark:brown', hue = "sex")
plt.show()
```

# creating boxplot for each of the numaric value provided in the dataset

```python
sns.boxplot(data=tip)
plt.show()
```
