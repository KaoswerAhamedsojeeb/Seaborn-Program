```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print(sns.get_dataset_names())
```

```python
data = sns.load_dataset("titanic")
data.head()
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(data=data,x="age",color="green", fill=True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(data=data,x="age",y = "fare",color="green", fill=True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.kdeplot(data=data,x="age",hue="class",color="green", fill=True)
plt.show()
```
