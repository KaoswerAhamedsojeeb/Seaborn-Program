```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
sns.regplot(data=df,x="sepal_length", y="sepal_width")
plt.show()
```

```python
sns.regplot(data=df,x="sepal_length", y="petal_length")
plt.show()
```

```python
sns.regplot(
    data=df,
    x="sepal_length",
    y="petal_length",
    color="purple",
    marker="x",
    scatter_kws={"s": 60},
    line_kws={"color": "black", "linewidth": 2}
)
```
