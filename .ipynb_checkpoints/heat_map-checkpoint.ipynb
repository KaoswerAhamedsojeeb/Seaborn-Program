```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('iris')
data.head()
```

```python
corr_data=data.corr(numeric_only=True)
```

```python
corr_data.head()
```

```python
sns.heatmap(data=corr_data, annot=True)
plt.show()
```

```python
sns.heatmap(corr_data, annot=True, cmap="coolwarm", linewidths=0.5)
plt.title(" Iris Feature Correlation")
plt.show()
```

```python
sns.heatmap(corr_data, annot=True, cmap="coolwarm", linewidths=0.5, fmt=".3f")
plt.title(" Iris Feature Correlation")
plt.show()
```
