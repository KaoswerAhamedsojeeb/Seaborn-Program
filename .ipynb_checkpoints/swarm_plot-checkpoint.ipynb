```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = sns.load_dataset("tips")
```

```python
data.head()
```

```python
sns.swarmplot(data=data,y="total_bill", x= "sex")
plt.grid(True)
plt.show()
```

```python
sns.swarmplot(data=data,y="total_bill", x= "sex",linewidth=1.2,edgecolor="green", color="black")
plt.grid(linestyle="--",alpha=0.5)
plt.show()
```

```python
sns.swarmplot(data=data,y="total_bill", x= "day",hue="sex",linewidth=1.2)
plt.grid(linestyle="--",alpha=0.5)
plt.show()
```

```python
sns.swarmplot(data=data,y="total_bill", x= "day",hue="sex",linewidth=1.2, marker="*")
plt.grid(linestyle="--",alpha=0.5)
plt.show()
```

```python
sns.swarmplot(data=data,y="total_bill", x= "day",linewidth=1.2,color="white")
sns.violinplot(data=data,y="total_bill", x= "day")
plt.grid(linestyle="--",alpha=0.5)
plt.show()
```
